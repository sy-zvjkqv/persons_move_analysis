```python
merged_Tweet_place = pd.read_csv('/home/is/shuntaro-o/dev/persons_move_analysis/data/202271month_per_hour_geotaged_adGeocode.csv',engine='python')
```

```python
Tokyo_tweets = merged_Tweet_place[merged_Tweet_place["code"]==5339]
```

```python
def UTM_cal(lat,lon):
    p,a=divmod(lat*60,40)
    q,b=divmod(a,5)
    r,c=divmod(b*60,30)
    s,d=divmod(c,15)
    t,e=divmod(d,7.5)

    u=str(lon-100)[0:2]
    f=lon-int(u)-100
    v,g=divmod(f*60,7.5)
    w,h=divmod(g*60,45)
    x,i=divmod(h,22.5)
    y,j=divmod(i,11.25)

    m=(s*2)+(x+1)
    n=(t*2)+(y+1)
    p=int(p)
    u=int(u)
    q=int(q)
    v=int(v)
    r=int(r)
    w=int(w)
    m=int(m)
    n=int(n)

    ans=str(p)+str(u)+str(q)+str(v)
    ans=int(ans)
    return(ans)
code_list=[]
for index, r in Tokyo_tweets.iterrows():
    ans=UTM_cal(r.UserPlase_latitude,r.UserPlase_longitude)
    code_list.append(ans)
Tokyo_tweets = Tokyo_tweets.reset_index(drop=True)
code_list=pd.Series(code_list)
Tokyo_tweets=pd.concat([Tokyo_tweets,code_list],axis=1)
Tokyo_tweets=Tokyo_tweets.rename(columns={0: 'code_Tokyo'})
```

```python
Tokyo_tweets=Tokyo_tweets.rename(columns={'code': 'code_nation'})
Tokyo_tweets=Tokyo_tweets.rename(columns={'code_Tokyo': 'code'})
```
