```python
import pandas as pd

UserPlase = pd.DataFrame()
files = glob.glob('/home/is/shuntaro-o/dev/disaster_analysis_Twitter/data/TweetPlaces_Japan7Cities/*.txt')
for file in files:
    # csvファイルを試験データを読込む
    tmp = pd.read_csv(file,sep=',', lineterminator='\n', header=None,names = ['place_id', 'city_name', 'object', 'country', 'UserPlase_longitude', 'UserPlase_latitude', 'UserPlase_sub_longitude', 'UserPlase_sub_latitude'])
    UserPlase = pd.concat([UserPlase,tmp],axis=0)
```

```python
UserPlase =UserPlase[~UserPlase.duplicated(subset='place_id')]
UserPlase = UserPlase.reset_index(drop=True)
UserTweets = UserTweets.reset_index(drop=True)
merged_Tweet_place = pd.merge(UserTweets , UserPlase, how='left', on = "place_id")
merged_Tweet_place = merged_Tweet_place.reset_index(drop=True)
merged_Tweet_place = merged_Tweet_place.dropna(subset=['UserPlase_latitude','UserPlase_longitude'])
```

```python
def UTM_cal(lat,lon):
    p,a=divmod(lat*60,40)
    q,b=divmod(a,5)
    r,c=divmod(b*60,30)
    s,d=divmod(c,15)
    t,e=divmod(d,7.5)

    u=str(lon-100)[0:2]
    f=lon-int(u)-100
    v,g=divmod(f*60,7.5)
    w,h=divmod(g*60,45)
    x,i=divmod(h,22.5)
    y,j=divmod(i,11.25)

    m=(s*2)+(x+1)
    n=(t*2)+(y+1)
    p=int(p)
    u=int(u)
    q=int(q)
    v=int(v)
    r=int(r)
    w=int(w)
    m=int(m)
    n=int(n)

    ans=str(p)+str(u)
    ans=int(ans)
    return(ans)
code_list=[]
for index, r in merged_Tweet_place.iterrows():
    ans=UTM_cal(r.UserPlase_latitude,r.UserPlase_longitude)
    code_list.append(ans)
merged_Tweet_place = merged_Tweet_place.reset_index(drop=True)
code_list=pd.Series(code_list)
merged_Tweet_place=pd.concat([merged_Tweet_place,code_list],axis=1)
merged_Tweet_place=merged_Tweet_place.rename(columns={0: 'code'})
```

```python
shuffled = merged_Tweet_place.sample(frac=1,random_state=0)
```

```python
count_codes = shuffled['code'].value_counts()
```

```python
count_codes_dict = count_codes.to_dict()
```

```python
count_codes_dict[5339]=30000
count_codes_dict[5235]=30000
count_codes_dict[5135]=30000
count_codes_dict[5236]=30000
count_codes_dict[5030]=30000
count_codes_dict[5439]=30000
```

```python
#アンダーサンプリングで学習用データ作成

from imblearn.under_sampling import RandomUnderSampler


# クラス1が全体の10％になるまでクラス0を減らす
under = RandomUnderSampler(sampling_strategy=count_codes_dict, random_state=0)

# 学習用データに反映
sampled,_ = under.fit_resample(merged_Tweet_place,merged_Tweet_place["code"])
```
